```python
import wandb

def setup_wandb_kaggle(project_name: str, entity: str, run_name: str = None, secret_name: str = "WANDB_API_KEY"):
    try:
        api_key = os.getenv(secret_name)
        wandb.login(key=api_key, relogin=True)
        print("‚úÖ Login a wandb riuscito con secret!")

        wandb.init(project=project_name, entity=entity, name=run_name)
        print(f"üöÄ Progetto '{project_name}' inizializzato nel team '{entity}'.")
    except KeyError:
        print(f"‚ùå Secret '{secret_name}' non trovato nei Kaggle secrets.")
    except Exception as e:
        print("‚ùå Errore nel setup wandb:", e)
```
